DTVC Coursework Google Trends to see consumer searches about wine 

Load Modules

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from datetime import datetime, timedelta
from pytrends.request import TrendReq


#Google Trends
pytrend = TrendReq()
pytrend.build_payload(kw_list = ["Buy Wines"],timeframe='2021-10-29 2022-10-29')
Buywine_trends = pytrend.interest_over_time()
Buywine_trends
# %%
Buywine_trends = pd.DataFrame(data = {'date': Buywine_trends.index.tolist(), 'search': Buywine_trends["Buy Wines"].tolist()})
Buywine_trends
# %%
#Scatterplot
sns.scatterplot( x = 'date',
             y = 'search',
             data = Buywine_trends,
             label = 'daily hits', color="blue")
plt.show()

# %%
#Adding a rolling average
Buywine_trends['rolling_avg' ] = Buywine_trends.search.rolling(7).mean()
Buywine_trends = Buywine_trends.fillna(0)
sns.scatterplot( x = 'date',
             y = 'search',
             data = Buywine_trends,
             label = 'daily hits', color="blue")
sns.lineplot( x = 'date',
             y = 'rolling_avg',
             data = Buywine_trends,
             label = 'rolling_avg', color="orange")
plt.show()

#%%
#Searches for price of wines in the UK
pytrend = TrendReq()
pytrend.build_payload(kw_list = ["Price of Wines"],timeframe='2021-10-29 2022-10-29')
Pricewine_trends = pytrend.interest_over_time()
Pricewine_trends
# %%
Pricewine_trends = pd.DataFrame(data = {'date': Pricewine_trends.index.tolist(), 'search': Pricewine_trends["Price of Wines"].tolist()})
Pricewine_trends
# %%
#Scatterplot
sns.scatterplot( x = 'date',
             y = 'search',
             data = Pricewine_trends,
             label = 'daily hits', color="blue")
plt.show()

# %%
#Adding a rolling average
Pricewine_trends['rolling_avg' ] = Pricewine_trends.search.rolling(7).mean()
Pricewine_trends = Pricewine_trends.fillna(0)
sns.scatterplot( x = 'date',
             y = 'search',
             data = Pricewine_trends,
             label = 'daily hits', color="blue")
sns.lineplot( x = 'date',
             y = 'rolling_avg',
             data = Pricewine_trends,
             label = 'rolling_avg', color="orange")
plt.show()
```
